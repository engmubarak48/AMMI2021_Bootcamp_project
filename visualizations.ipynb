```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
pd.options.display.max_columns=None
pd.options.display.max_rows=None
%matplotlib inline
```

```python
df = pd.read_csv("data/train_data.txt",header=None)
```

```python
df.head()
```

```python
col_names=['ID', 'f1', 'f2', 'f3', 'f4', 'f5', 'f6', 'f7', 'f8', 'f9', 'f10', 'f11', 'Class']
df.columns=col_names
```

```python
df.head()
```

```python
df['Class'].value_counts().plot(kind='bar')
```

```python
df2 =df.drop(['ID','Class'],axis=1)
```

```python
def plot_corr(df):
    return df.corr()
```

```python
corre=plot_corr(df2)
```

```python
corre
```

```python
fig,ax = plt.subplots(4,3,sharex=True,figsize=(20,12))
X=df['f1']
y=['f2','f3','f4','f5','f6','f7','f8','f9','f10','f11']
for i in range(4):
    for j in range(3):
        index = i+j+2*i
        if index < len(y):
            Y=df[y[index]]

            scatter = ax[i,j].scatter(X,Y,c=df['Class'])
            leg = ax[i,j].legend(*scatter.legend_elements(),loc="upper right")
            ax[i,j].add_artist(leg)
        else:
            break;
```
